# Gender Distribution

```python
plt.figure(figsize=(5,5)) #to adjust the size of the graph
ax=sns.countplot(data=df, x="Gender") #countplot---> it counts the value
ax.bar_label(ax.containers[0]) # to plot bar graph
plt.title("Gender Distribution")
plt.show()
```

#from the above chart we have analysed that:
#the number of females in the data is more than the number of males.

```python
gp = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'}) #to analysis parent edu
print(gp)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gp,annot=True)#heatmap to better presentation of value. annot=True--->to show the value.
plt.title("parents education impact")
plt.show()
```

```python
#from the above chart we have conclude that the education of the parents have good impact.
```

```python
gp1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'}) #to analysis ParentMaritalStatus
print(gp1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gp1,annot=True) #heatmap to better presentation of value. annot=True--->to show the value.
plt.title("Parents martial impact")
plt.show()
```

```python
#from the above chart we have conclude that the education of the parents have no impact or neglible impact.
```

# distribution of Ethenic Group

```python
plt.figure(figsize=(5,4))
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()
l=["groupA","groupB","groupC","groupD","groupE",]
mlist= [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"] ,groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels= l, autopct="%1.2f%%")
plt.title("Distribution of ethnic groups")
plt.show()
```
